## 데이터 프레임 생성

(1) 하나의 열이 되는 데이터를 리스트나 일차원 배열을 준비한다.

(2) 이 각각의 열에 대한 이름(라벨)을 키로 가지는 딕셔너리를 만든다.

(3) 이 데이터를 DataFrame 클래스 생성자에 넣는다. 동시에 열방향 인덱스는 columns 인수로, 행방향 인덱스는 index 인수로 지정한다.

```python
import pandas as pd
data = {
    "2015": [9904312, 3448737, 2890451, 2466052],
    "2010": [9631482, 3393191, 2632035, 2431774],
    "2005": [9762546, 3512547, 2517680, 2456016],
    "2000": [9853972, 3655437, 2466338, 2473990],
    "지역": ["수도권", "경상권", "수도권", "경상권"],
    "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
}
columns = ["지역", "2015", "2010", "2005", "2000", "2010-2015 증가율"] #열의 이름
index = ["서울", "부산", "인천", "대구"] #행방향 인덱스
df = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
#데이터만 접근하기 : value 속성 사용
df.values
```

```python
#열방향 인덱스 : columns 사용
#행방향 인덱스 : index 사용

df.columns
```

```python
df.index
```

```python
#열방향 인덱스와 행방향 인덱스에 이름 붙이기
df.index.name = '도시'
df.columns.name = '특성'
df
```

## 열 데이터 갱신, 추가, 삭제
- 추가 : df[추가할 열] = 데이터
- 삭제 : del[삭제할 열]

```python
#2010-2015 증가율 열의 데이터 갱신
df["2010-2015 증가율"] = df["2010-2015 증가율"]*100
df
```

```python
#2005-2010 증가율이라는 열 추가
df["2005-2010 증가율"] = ((df["2010"] - df["2005"]) / df["2005"]*100).round(2)
df
```

```python
#2010-2015 증가율이라는 이름의 열 삭제
del df["2010-2015 증가율"]
df
```

## 행 인덱싱 방법
 - 인덱스(loc)
 - 행번호(iloc)
 - 슬라이싱

#✨행단위 데이터 추출 : 인덱스/ 행 번호 사용
### - loc속성 = 인덱스기준 : 행 맨 앞에 있는 숫자, 언제든지 변할 수 있고 숫자가 아닌 문자열을 사용할 수도 있다. 
  (e.g. : first, second, third와 같은 문자열로도 지정 가능)
### - iloc속성 = 행 번호 : 정수만으로 데어터를 조회하거나 추출 가능, 실제 데이터프레임에서는 확인할 수 없음


```python
df.loc["부산"]
```

##iloc 속성으로 행 데이터 추출 
 print(df.iloc[행번호])
   숫자(정수형)만 가능

```python
df.iloc[1] 
```

```python
df.iloc[[0, 2, 3]] #여러 행을 출력하려면 데이터의 인덱스, 행번호를 리스트에 담아 출력
```

### - 슬라이싱 구문 : df.loc[시작하는 인덱스 : 끝나는 인덱스, ['열이름']]
  (e.g. : df.loc[0:3, ['서울', '경기'], df.iloc[0:3, [0, 2]) #iloc는 정수형만 가능

## 연습문제 4.1.3

```python
data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df2 = pd.DataFrame(data, index=index, columns=columns)
df
```

```python
#(1) 모든 학생의 수학 점수를 시리즈로 나타낸다.
df2["수학"]
```

```python
#(2) 모든 학생의 국어와 영어 점수를 데이터프레임으로 나타낸다.
df2[["국어", "영어"]]
```

```python
#(3) 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
df2["평균"] = (df2["국어"] + df2["수학"] + df2["영어"]) / len(df2.columns)
df2
```

```python
#(4) 방자의 영어 점수를 80점으로 수정하고 평균점수도 다시 계산한다.
df2["영어"]["방자"] = 80
df2["평균"] = (df2["국어"] + df2["수학"] + df2["영어"]) / len(df2.columns)
df2
```

```python
#(5) 춘향의 점수를 데이터프레임으로 나타낸다.
df2.loc[["춘향"]]
```

```python
#(6) 향단의 점수를 시리즈로 나타낸다.
df2.loc["향단"]
```
