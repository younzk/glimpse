##apply 변환
- 행이나 열 단위로 더 복잡한 처리를 하고 싶을 때 사용하는 메서드
- 인수로 행 또는 열을 받는 함수를 apply 메서드의 인수로 넣으면 각 열이나 행을 반복하여 그 함수에 적용시킨다.

- 행으로 적용하고 싶을 경우 axis = 1을 사용
- value_counts : 각 열에 대해 어떤 값이 얼마나 사용되었는지 알아볼 때

```python
df3 = pd.DataFrame({
    'A' : [1, 3, 4, 3, 4],
    'B' : [2, 3, 1, 2, 3], 
    'C' : [1, 5, 2, 4, 4, ]
})
df3
```

###각 열의 최댓값과 최솟값의 차이를 구하고 싶을 때

```python
df3.apply(lambda x : x.max() - x.min())
```

### 각 열에서 어떤 값이 얼마나 사용되었는지 알고 싶을 때

```python
df3.apply(pd.value_counts)
```

```python
titanic["adult/child"] = titanic.apply(lambda r : "adult" if r.age >= 20 else "child", axis = 1)
titanic.tail()
```

##실수 값을 카테고리 값으로 변환
- bins : 카테고리를 나누는 기준 값
- labels : 카테고리 이름
- cut : 실수 값의 경계선을 지정하는 경우
- qcut : 갯수가 똑같은 구간으로 나누는 경우

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 101]
```

```python
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(ages, bins, labels=labels) 
cats

#ages값을 카테고리로 바꾸는데, bins의 숫자가 경계값이 됨
#1~20은 미성년자
#20~30은 청년
#30~50은 중년
#50~70은 장년
#70~100은 노년
#구간에 포함되지 않는 값은 NaN으로
```

```python
cats.categories #카테고리 라벨 문자열
```

```python
cats.codes #정수로 인코딩한 카테고리 값
```

```python
df4 = pd.DataFrame(ages, columns=["ages"])
df4["age_cat"] = pd.cut(df4.ages, bins, labels=labels)
df4
```

```python
df4.age_cat.astype(str) + df4.ages.astype(str)
```

###qcut 활용하기
