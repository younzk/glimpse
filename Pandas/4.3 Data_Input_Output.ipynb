#4.3 Data Input / Output

## %%writefile 명령

csv 파일을 %%writefile 매직명령으로 만들기
셀에 서술한 내용대로 텍스트 파일을 만드는 명령이다.

```python
%%writefile sample1.csv
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

## CSV 파일 입력

```python
import pandas as pd
pd.read_csv('sample1.csv') #위에서 생성한 csv 파일을 읽기
#인덱스를 지정하지 않은 경우 0부터 시작하게 된다.
```

```python
%%writefile sample2.csv
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
#인덱스를 names 인수로 설정할 수 있다.
pd.read_csv('sample2.csv', names=['인덱스1', '인덱스2', '인덱스3'])
```

```python
#테이블 내의 특정 열을 행 인덱스로 지정하고 싶은 경우 index_col을 사용
pd.read_csv('sample2.csv', index_col = '1')
```

```python
#건너뛰어야 할 행이 있는 경우 skiprows를 사용
%%writefile sample4.txt
파일 제목: sample4.txt
데이터 포맷의 설명:
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample4.txt', skiprows=[0, 1])
```

```python
#특정한 값을 NaN으로 취급하고 싶으면 na_values인수에 NaN값으로 취급할 값을 넣는다.
%%writefile sample5.csv
c1, c2, c3
1, 1.11, one
2, , two
누락, 3.33, three
```

```python
df = pd.read_csv('sample5.csv', na_values=['누락']) #"누락"이라는 값을 NaN으로 처리
df
```

## CSV 파일 출력 : to_csv 메서드 사용

```python
df.to_csv('sample6.csv')
```
