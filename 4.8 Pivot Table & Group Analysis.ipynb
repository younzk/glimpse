##Pivot Table
- Pivot Table이란 데이터 열 중에서 두 개의 열을 각각 행, 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것
-df1.pivot("행 인덱스로 사용할 열 이름", "열 인덱스로 사용할 열 이름", "데이터로 사용할 열 이름")

- 판다스는 지정된 두 열을 각각 행 인덱스와 열 인덱스로 바꾸고 행 인덱스의 라벨 값이 첫번째 키의 값과 같고 두번째 키의 값과 같은 데이터를 찾아서 해당 칸에 넣는다. 만약 데이터가 존재하지 않으면 NaN값을 넣는다.

```python
import pandas as pd
import numpy as np
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
columns = ["도시", "연도", "인구", "지역"]
df1 = pd.DataFrame(data, columns=columns)
df1
```

```python
df1.pivot("도시", "연도", "인구") #행인덱스 : 도시, 열인덱스 : 연도, 데이터 : 인구
```

피봇테이블은 set_index와 unstack 명령을 사용해서 만들 수도 있다.
- set_index : 해당 데이터열을 인덱스로 설정
- unstack : 행 인덱스 -> 열 인덱스

```python
df1.set_index(["도시", "연도"])[["인구"]]
```

```python
df1.set_index(["도시", "연도"])[["인구"]].unstack()
```

```python
df1.set_index(["도시", "연도"])[["인구"]].unstack().unstack()
```

```python
df1.pivot(["지역", "도시"], "연도", "인구")
```

행 인덱스와 열 인덱스는 데이터를 찾는 key역할을 하므로 키 값으로 데이터가 하나면 찾아져야 한다.
행 인덱스와 열 인덱스 조건을 만족하는 데이터가 2개 이상인 경우에는 에러가 발생한다.
예를 들어 위에서 "지역", "연도"를 키로 하면 "수도권", "2015"에 해당하는 값이 두 개 이상이므로 에러가 발생한다.

```python
try:
  df1.pivot("지역", "연도", "인구")
except ValueError as e:
  print("ValueError: ", e)
```

##Grourp Analysis(그룹 분석)
- 키가 지정하는 조건에 맞는 데이터가 하나 이상이라 데이터 그룹을 이루는 경우 사용
- groupby 메서드 사용

##그룹 연산 메서드
- size, count : 그룹 데이터의 종류
- mean, median, min, max : 그룹 데이터의 평균, 중앙값, 최소, 최대
- sum, prod, std, var, quantile : 그룹 데이터의 합계, 곱, 표준편차, 분산, 사분위수
- first, last : 그룹 데이터 중 가장 첫번째 데이터와 가장 나중 데이터
- agg, aggregate : 원하는 그룹 연산이 없는 경우 함수를 만들고 이 함수를 agg에 전달, 여러가지 그룹연산을 동시에 하고 싶은 경우 함수 이름 문자열의 리스트 전달
- describe : 여러 개의 값을 데이터 프레임으로 구하기
- apply : describe처럼 하나의 대표값이 아닌 데이터프레임을 출력하지만 원하는 그룹 연산이 없는 경우 사용
- transform : 그룹 대표값을 만드는게 아니라 그룹별 계산을 통해 데이터 자체를 변형

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

key1의 값에 따른 data1의 평균
1. groupby명령을 사용해 그룹 A와 그룹 B로 구분한 그룹데이터를 만든다.

```python
groups = df2.groupby(df2.key1).mean()
groups
```

GroupBy 클래스 객체에는 각 그룹의 인덱스를 저장한 groups 속성이 있다.

A그룹과 B그룹의 합계를 위해 sum이라는 그룹 연산을 한다.

```python
groups.sum()
```

GroupBy 클래스가 필요 없다면 groupby메서드와 그룹연산 메서드를 연속 호출한다.

```python
df2.data1.groupby(df2.key1).sum() #data1에 대해서 key1을 기준으로 그룹 -> sum 연산
```

```python
df2.groupby(df2.key1)["data1"].sum() #df2전체를 key1기준으로 그룹 -> 그중 data1 만 -> sum 연산 ('GroupBy) 클래스 객체에서 data1만 선택하여 분석
```

```python
df2.groupby(df2.key1).sum()["data1"] #전체 데이터를 분석 후 data1만 선택
```

복합 키(key1, key2) 값에 따른 data1의 합계 구하기
- 키가 복수인 경우 리스트 사용

```python
df2.data1.groupby([df2.key1, df2.key2]).sum()
```

```python
df2.data1.groupby([df2["key1"], df2["key2"]]).sum().unstack("key2") #key2(행인덱스)를 열 인덱스로 바꿔준다
```

150송이의 붓꽃(iris)에 대해 붓꽃 종(species)별로 꽃잎 길이(sepal_length), 꽃잎 폭(sepal_width)을 측정한 데이터 불러오기

```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris.head(10)
```

각 붓꽃 종별로 가장 큰 값과 가장 작은 값을 비율 구하기
- 메서드가 없으므로 직접 만든 후 agg 메서드를 적용한다.

```python
def peak_to_peak_ratio(x):
  return x.max() / x.min()

iris.groupby(iris.species).agg(peak_to_peak_ratio) #species로 그룹한 데이터에서 가장 큰 값(max)와 가장 작은 값(min)의 비율
```

describe 메서드를 사용하면 다양한 기술 통계(descriptive statistics)값을 한 번에 구한다.

```python
iris.groupby(iris.species).describe().T 
```

apply 메서드를 사용하면 describe 메서드처럼 하나의 그룹에 대해 하나의 대표값(스칼라 값)을 구하는 게 아니라 데이터프레임을 만들 수 있다.  
 예를 들어 다음처럼 각 붓꽃 종별로 가장 꽃잎 길이(petal length)가 큰 3개의 데이터를 뽑아낼 수도 있다.

```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3] #sort_values() : 데이터값 기준으로 정렬,  ascending = False :큰 수에서 작은 수로 반대 정렬, by = 기준이 되는 열

k = iris.groupby(iris.species).apply(top3_petal_length)
k
```

```python
k["sepal_length"]["versicolor"]
```

transform 메서드는 그룹별 대표값을 만드는 것이 아니라 그룹별 계산을 통해 데이터프레임 자체를 변화시킨다.   
따라서 만들어진 데이터프레임의 크기는 원래 데이터프레임과 같다.   
예를 들어 다음처럼 각 붓꽃 꽃잎길이가 해당 종 내에서 대/중/소 어느 것에 해당되는지에 대한 데이터프레임을 만들 수도 있다.

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["소", "중", "대"]).astype(str) #s에 들어오는 값을 3가지 카테고리로 나누기 , astype으로 전체 자료형을 바꾼다.


iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut)
iris[["petal_length", "petal_length_class"]].tail(10)
```

##pivot_table
- pivot명령과 groupby 명령의 중간 성격을 가짐
- groupby처럼 그룹 분석을 하지만 최종적으로 pivot처럼 피봇테이블을 만든다.
- groupby 명령의 결과에 unstack을 자동 적용하여 2차원 형태로 변형한다.

* pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All')

 - data: 분석할 데이터프레임 (메서드일 때는 필요하지 않음)

 - values: 분석할 데이터프레임에서 분석할 열

 - index: 행 인덱스로 들어갈 키 열 또는 키 열의 리스트

 - columns: 열 인덱스로 들어갈 키 열 또는 키 열의 리스트

 - aggfunc: 분석 메서드

 - fill_value: NaN 대체 값

 - margins: 모든 데이터를 분석한 결과를 오른쪽과 아래에 붙일지 여부

 - margins_name: 마진 열(행)의 이름

```python
df1.pivot_table("인구", "도시", "연도")
```

margins = True 인수 : aggfunc으로 주어진 분석방법을 해당 열의 모든 데이터, 해당 행의 모든 데이터, 그리고 전체 데이터에 대해 적용한 결과를 보여준다. aggfunc이 주어지지 않았으면 평균을 계산한다.

아래 결과에서 가장 오른쪽 합계 열의 첫번째 값 3583992은 모든 부산 인구 데이터의 평균, 두번째 값 9749333은 모든 서울 인구 데이터의 평균이다. 가장 아래의 합계 행의 첫번째 값은 2005년 데이터의 평균값, 두번째 값은 2010년 데이터의 평균값이다. 가장 오른쪽 아래의 값 7283197는 전체 데이터의 평균값이다. 다음 계산을 통해 이를 확인할 수 있다.

```python
df1.pivot_table("인구", "도시", "연도", margins=True, margins_name="합계") #aggfunc이 없으므로 평균
```

따라서 금요일 점심, 1인 테이블에서 가장 많은 팁을 받을 수 있다.
- 어떤 요인이 크게 작용하는지 판단하는 방법?

```python
def peak_to_peak(x):
  return x.max() - x.min() #peak_to_peak : 가장 많은 팁과 적은 팁의 차이

tips.groupby(["sex", "smoker"])[["tip"]].agg(peak_to_peak)
```

- 여러가지 그룹 연산을 동시에 하고 싶을 때 : 리스트를 사용
- 데이터 열마다 다른 연산을 하고 싶다면 딕셔너리로 넣는다. 
{열 라벨 : 연산 이름 또는 함수}

```python
#여러가지 그룹 연산을 동시에 하고 싶을 때 : 리스트를 사용
tips.groupby(["sex", "smoker"]).agg(["mean", peak_to_peak])[["total_bill"]]
```

```python
tips.groupby(["sex", "smoker"]).agg(
    {'tip_pct' : 'mean', 'total_bill' : peak_to_peak} #tip_pct열은 평균값, total_bill은 peak_to_peak연산
)
```

```python
tips.pivot_table(['tip_pct', 'size'], ['sex', 'day'], 'smoker') #데이터 내용은 tip_pct와 size, 행데이터 인덱스 : sex, day, 열데이터 인덱스 : smoker
```

```python
tips.pivot_table('size', ['time', 'sex', 'smoker'], 'day', 
                 aggfunc = 'sum', fill_value = 0)
```
